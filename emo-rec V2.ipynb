```python
import numpy as np
import pandas as pd
import random
import os
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout,
    BatchNormalization, GlobalAveragePooling2D
)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.regularizers import l2
import librosa

# RAVDESS emotion mapping
ravdess_emotion_map = {
    '01': 'neutral', '02': 'calm', '03': 'happy', '04': 'sad',
    '05': 'angry', '06': 'fearful', '07': 'disgust', '08': 'surprised'
}

# TESS emotion mapping 
tess_emotion_map = {
    'neutral': 'neutral',
    'happy': 'happy',
    'sad': 'sad',
    'angry': 'angry',
    'fear': 'fearful',
    'disgust': 'disgust',
    'ps': 'surprised'  
}

# Dataset paths
try:
    # RAVDESS dataset path
    ravdess_root_dir = "/kaggle/input/audio-data"
    if not os.path.exists(ravdess_root_dir):
        print("Kaggle RAVDESS path not found. Trying local paths...")
        if os.path.exists("./audio-data"):
            ravdess_root_dir = "./audio-data"
        elif os.path.exists("./input/audio-data"):
            ravdess_root_dir = "./input/audio-data"
        else:
            ravdess_root_dir = "/kaggle/input/audio-data"
            if not os.path.exists(ravdess_root_dir):
                print(f"Warning: RAVDESS dataset directory {ravdess_root_dir} not found.")

    # TESS dataset path
    tess_root_dir = "/kaggle/input/toronto-emotional-speech-set-tess"
    if not os.path.exists(tess_root_dir):
        print("Kaggle TESS path not found. Trying local paths...")
        if os.path.exists("./toronto-emotional-speech-set-tess"):
            tess_root_dir = "./toronto-emotional-speech-set-tess"
        elif os.path.exists("./input/toronto-emotional-speech-set-tess"):
            tess_root_dir = "./input/toronto-emotional-speech-set-tess"
        else:
            print(f"Warning: TESS dataset directory {tess_root_dir} not found.")

except Exception as e:
    print(f"Error setting dataset paths: {e}")
    ravdess_root_dir = "./audio-data"
    tess_root_dir = "./toronto-emotional-speech-set-tess"

# Process RAVDESS dataset
print("Processing RAVDESS dataset...")
ravdess_data = []
if os.path.exists(ravdess_root_dir):
    for subdir, _, files in os.walk(ravdess_root_dir):
        for file in files:
            if file.endswith(".wav"):
                try:
                    parts = file.split("-")
                    emotion_code = parts[2]
                    emotion = ravdess_emotion_map[emotion_code]
                    filepath = os.path.join(subdir, file)
                    ravdess_data.append((filepath, emotion, 'RAVDESS'))
                except (KeyError, IndexError):
                    print(f"Could not process RAVDESS file: {file}")
    print(f"Found {len(ravdess_data)} RAVDESS files")
else:
    print("RAVDESS directory not found!")

# Process TESS dataset
print("Processing TESS dataset...")
tess_data = []
if os.path.exists(tess_root_dir):
    for subdir, _, files in os.walk(tess_root_dir):
        for file in files:
            if file.endswith(".wav"):
                try:
                   
                    filename_lower = file.lower()
                    emotion = None

                    # Extract emotion from filename
                    for tess_key, mapped_emotion in tess_emotion_map.items():
                        if tess_key in filename_lower:
                            emotion = mapped_emotion
                            break

                    if emotion is None:
                        if 'neutral' in filename_lower:
                            emotion = 'neutral'
                        elif 'happy' in filename_lower:
                            emotion = 'happy'
                        elif 'sad' in filename_lower:
                            emotion = 'sad'
                        elif 'angry' in filename_lower:
                            emotion = 'angry'
                        elif 'fear' in filename_lower:
                            emotion = 'fearful'
                        elif 'disgust' in filename_lower:
                            emotion = 'disgust'
                        elif 'surprise' in filename_lower or 'ps' in filename_lower:
                            emotion = 'surprised'

                    if emotion:
                        filepath = os.path.join(subdir, file)
                        tess_data.append((filepath, emotion, 'TESS'))
                    else:
                        print(f"Could not determine emotion for TESS file: {file}")

                except Exception as e:
                    print(f"Error processing TESS file {file}: {e}")
    print(f"Found {len(tess_data)} TESS files")
else:
    print("TESS directory not found!")

# Combine datasets
all_data = ravdess_data + tess_data
print(f"Total combined dataset size: {len(all_data)} files")

# Create combined dataframe
df = pd.DataFrame(all_data, columns=["path", "emotion", "dataset"])

# DROP 'calm' EMOTION 
print("\nOriginal emotion distribution:")
print(df['emotion'].value_counts())

print("\nDropping 'calm' emotion to improve balance and focus...")
df = df[df['emotion'] != 'calm'].reset_index(drop=True)

print("\nNew emotion distribution after dropping 'calm':")
print(df['emotion'].value_counts())


# Print dataset statistics
print("\nDataset Statistics (after dropping 'calm'):")
print("By Dataset:")
print(df['dataset'].value_counts())
print("\nBy Emotion:")
print(df['emotion'].value_counts())
print("\nBy Dataset and Emotion:")
print(df.groupby(['dataset', 'emotion']).size().unstack(fill_value=0))


def add_noise(y, noise_factor_range=(0.002, 0.008)):
    noise_factor = random.uniform(noise_factor_range[0], noise_factor_range[1])
    noise = np.random.randn(len(y))
    return (y + noise_factor * noise).astype(type(y[0]))

def pitch_shift(y, sr, n_steps_range=(-3, 3)):
    n_steps = random.randint(n_steps_range[0], n_steps_range[1])
    if n_steps == 0: return y
    return librosa.effects.pitch_shift(y=y, sr=sr, n_steps=n_steps)

def time_stretch(y, rate_range=(0.8, 1.2)):
    rate = random.uniform(rate_range[0], rate_range[1])
    if rate == 1.0: return y
    return librosa.effects.time_stretch(y=y, rate=rate)

def spec_augment(spec, time_masking_para=40, freq_masking_para=30, num_time_masks=2, num_freq_masks=2):
    spec_copy = spec.copy()
    n_mels, n_steps = spec_copy.shape
    for _ in range(num_freq_masks):
        f = random.randrange(0, freq_masking_para)
        f0 = random.randrange(0, n_mels - f)
        if f > 0:
            spec_copy[f0:f0+f, :] = 0
    for _ in range(num_time_masks):
        t = random.randrange(0, time_masking_para)
        t0 = random.randrange(0, n_steps - t)
        if t > 0:
            spec_copy[:, t0:t0+t] = 0
    return spec_copy

def extract_delta_features(y, sr, n_mels=128, n_mfcc=40, max_len=250):
    mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=n_mels)
    log_mel = librosa.power_to_db(mel)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    mfcc_delta = librosa.feature.delta(mfcc)
    mfcc_delta2 = librosa.feature.delta(mfcc, order=2)
    features = np.vstack((log_mel, mfcc, mfcc_delta, mfcc_delta2))

    # Pad or truncate
    if features.shape[1] < max_len:
        features = np.pad(features, ((0, 0), (0, max_len - features.shape[1])), mode='constant')
    else:
        features = features[:, :max_len]

    # Per-instance standardization
    if features.size > 0:
        features = (features - np.mean(features)) / (np.std(features) + 1e-6)

    return features


le = LabelEncoder()
df['emotion_encoded'] = le.fit_transform(df['emotion'])


train_df, test_df = train_test_split(
    df,
    test_size=0.2,
    random_state=42,
    stratify=df['emotion']
)

print(f"\nTrain set size: {len(train_df)}")
print(f"Test set size: {len(test_df)}")
print("\nTrain set distribution:")
print(train_df.groupby(['dataset', 'emotion']).size().unstack(fill_value=0))
print("\nTest set distribution:")
print(test_df.groupby(['dataset', 'emotion']).size().unstack(fill_value=0))

X_train, y_train, X_test, y_test = [], [], [], []
SR = 16000

print("\nProcessing and augmenting training data with delta features...")
processed_count = 0
for idx, row in train_df.iterrows():
    try:
        y_audio, sr = librosa.load(row['path'], sr=SR)
        if len(y_audio) == 0:
            print(f"Warning: Empty audio file skipped: {row['path']}")
            continue

        if row['dataset'] == 'TESS':
            audios = [y_audio, add_noise(y_audio)]
        else:
            audios = [y_audio, add_noise(y_audio), pitch_shift(y_audio, sr), time_stretch(y_audio)]

        for audio in audios:
            if len(audio) == 0:
                continue
            features = extract_delta_features(audio, sr)
            augmented_features = spec_augment(features)
            X_train.append(augmented_features)
            y_train.append(row['emotion_encoded'])

        processed_count += 1
        if processed_count % 100 == 0:
            print(f"Processed {processed_count}/{len(train_df)} training files...")

    except Exception as e:
        print(f"Error processing training file {row['path']}: {e}")

print(f"\nProcessing test data with delta features...")
processed_count = 0
for idx, row in test_df.iterrows():
    try:
        y_audio, sr = librosa.load(row['path'], sr=SR)
        if len(y_audio) == 0:
            print(f"Warning: Empty audio file skipped: {row['path']}")
            continue
        features = extract_delta_features(y_audio, sr)
        X_test.append(features)
        y_test.append(row['emotion_encoded'])

        processed_count += 1
        if processed_count % 100 == 0:
            print(f"Processed {processed_count}/{len(test_df)} test files...")

    except Exception as e:
        print(f"Error processing test file {row['path']}: {e}")

X_train = np.array(X_train)[..., np.newaxis]
y_train = np.array(y_train)
X_test = np.array(X_test)[..., np.newaxis]
y_test = np.array(y_test)

print(f"\nFinal dataset shapes:")
print(f"X_train: {X_train.shape}")
print(f"y_train: {y_train.shape}")
print(f"X_test: {X_test.shape}")
print(f"y_test: {y_test.shape}")

if X_train.size == 0 or X_test.size == 0:
    raise ValueError("Training or testing data is empty. Please check data loading and preprocessing.")


class_weights = class_weight.compute_class_weight(
    'balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights_dict = dict(enumerate(class_weights))
print(f"\nClass weights: {class_weights_dict}")

strategy = tf.distribute.MirroredStrategy()
print(f'Number of devices: {strategy.num_replicas_in_sync}')

with strategy.scope():
    def build_enhanced_cnn(input_shape, num_classes):
        inp = Input(shape=input_shape)
        l2_strength = 0.001  

        # Block 1
        x = Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(l2_strength))(inp)
        x = BatchNormalization()(x)
        x = Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(l2_strength))(x)
        x = BatchNormalization()(x)
        x = MaxPooling2D((2, 2))(x)
        x = Dropout(0.2)(x)

        # Block 2
        x = Conv2D(128, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(l2_strength))(x)
        x = BatchNormalization()(x)
        x = Conv2D(128, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(l2_strength))(x)
        x = BatchNormalization()(x)
        x = MaxPooling2D((2, 2))(x)
        x = Dropout(0.25)(x)

        # Block 3
        x = Conv2D(256, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(l2_strength))(x)
        x = BatchNormalization()(x)
        x = Conv2D(256, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(l2_strength))(x)
        x = BatchNormalization()(x)
        x = MaxPooling2D((2, 3))(x)
        x = Dropout(0.3)(x)

        # Block 4 
        x = Conv2D(512, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(l2_strength))(x)
        x = BatchNormalization()(x)
        x = MaxPooling2D((2, 2))(x)
        x = Dropout(0.35)(x)

        x = GlobalAveragePooling2D()(x)

        # Classifier
        x = Dense(512, activation='relu', kernel_regularizer=l2(l2_strength))(x)
        x = BatchNormalization()(x)
        x = Dropout(0.4)(x)

        x = Dense(256, activation='relu', kernel_regularizer=l2(l2_strength))(x)
        x = BatchNormalization()(x)
        x = Dropout(0.4)(x)

        x = Dense(num_classes, activation='softmax')(x)

        model = Model(inputs=inp, outputs=x)
        optimizer = Adam(learning_rate=0.0008)  
        model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
        return model

    # Build the model
    input_shape = X_train.shape[1:]
    num_classes = len(le.classes_)
    model = build_enhanced_cnn(input_shape, num_classes)
    model.summary()

#training
BATCH_SIZE = 32 * strategy.num_replicas_in_sync

checkpoint = ModelCheckpoint("best_model_fused_dataset_no_calm.keras", save_best_only=True, monitor="val_accuracy", mode="max", verbose=1)
early_stop = EarlyStopping(monitor='val_loss', patience=25, restore_best_weights=True, verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=10, min_lr=1e-7, verbose=1)

history = model.fit(
    X_train, y_train,
    epochs=150,
    validation_data=(X_test, y_test),
    callbacks=[early_stop, reduce_lr, checkpoint],
    batch_size=BATCH_SIZE,
    class_weight=class_weights_dict,
    verbose=1
)

#evaluation
print("\n--- Final Model Evaluation on Test Set (using best saved weights) ---")
model.load_weights("best_model_fused_dataset_no_calm.keras")
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy*100:.2f}%")

y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_, zero_division=0))

# Detailed analysis by dataset
test_df_reset = test_df.reset_index(drop=True)
test_predictions = []
for i, pred in enumerate(y_pred):
    test_predictions.append({
        'true_emotion': le.inverse_transform([y_test[i]])[0],
        'pred_emotion': le.inverse_transform([pred])[0],
        'dataset': test_df_reset.iloc[i]['dataset'],
        'correct': y_test[i] == pred
    })

pred_df = pd.DataFrame(test_predictions)
print("\nAccuracy by Dataset:")
dataset_accuracy = pred_df.groupby('dataset')['correct'].mean()
print(dataset_accuracy)

print("\nConfusion Matrix by Dataset:")
for dataset in pred_df['dataset'].unique():
    dataset_data = pred_df[pred_df['dataset'] == dataset]
    print(f"\n{dataset} Dataset:")
    print(f"Accuracy: {dataset_data['correct'].mean():.3f}")
    print(f"Sample size: {len(dataset_data)}")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(12, 10))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Confusion Matrix (Fused RAVDESS + TESS Dataset - No Calm)')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Training history plots
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy (Fused Dataset - No Calm)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss (Fused Dataset - No Calm)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 3)
dataset_accuracy.plot(kind='bar')
plt.title('Test Accuracy by Dataset')
plt.ylabel('Accuracy')
plt.xlabel('Dataset')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nModel training completed!")
print(f"Total training samples: {len(X_train)}")
print(f"Total test samples: {len(X_test)}")
print(f"Final test accuracy: {test_accuracy*100:.2f}%")
```
